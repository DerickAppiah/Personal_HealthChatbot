```python
from langchain.document_loaders import PyPDFLoader, DirectoryLoader #PyPDFLoader because im dealing with pdf, DirectoryLoader bcs im using directory
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
#extract data from pdf , *.pdf means all pdf in the directory
def load_pdf_file(data):
    loader= DirectoryLoader(data,
                            glob="*.pdf",
                            loader_cls=PyPDFLoader)

    documents=loader.load()

    return documents
```

```python
extracted_data=load_pdf_file(data='Data/')
```

# Now we we will start with creating chuncks 

```python
def text_split(extracted_data):
    text_splitter=RecursiveCharacterTextSplitter(chunk_size=500, chunk_overlap=20)
    text_chunks=text_splitter.split_documents(extracted_data)
    return text_chunks
```

```python
#length of text chunks
text_chunks=text_split(extracted_data)
print("Length of Text Chunks", len(text_chunks))
```

# Using embedding model to perform vector embedding so we can store the knowledge base in pinecon vector db

### we will download a model from hugging face 

```python
from langchain.embeddings import HuggingFaceEmbeddings
```

```python
def download_hugging_face_embeddings():
    embeddings=HuggingFaceEmbeddings(model_name='sentence-transformers/all-MiniLM-L6-v2')
    return embeddings

#we downloaded the model from hugging face
```

```python
embeddings = download_hugging_face_embeddings() #model downloading
```

```python
# pip install sentence-transformers
```

```python
#converting sentance to vector embedding using the hugging face model (testing)
query_result = embeddings.embed_query("Hello world")
print("Length", len(query_result))
```
